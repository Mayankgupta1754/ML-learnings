```python
#pichart
data['Pclass'].value_counts().plot(kind='pie', autopct='%0.2f')
```

```python
#cumulative frequency
a=data['Pclass'].value_counts()
a
```

```python
#adding percentages
b=((a/a.sum())*100).values
```

```python
b
```

```python
c=[]
cf=0
for i in b:
    cf=cf+i
    c.append(cf)
```

```python
c
```

```python
plt.plot(c, marker='D')
```
